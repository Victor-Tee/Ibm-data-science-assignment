```python
# To read a file and get the name
f = open('30days_Excel.txt', 'r')
print(f.name)
f.close()
```

```python
#to get the mode
f = open('30days_Excel.txt', 'r')
print(f.mode)
f.close()
```

```python
# how to open a file using a contact manager
with open('30days_Excel.txt', 'r') as f:
    f_contents = f.read()
    print(f_contents) #this read the contents of the file


```

```python

with open('30days_Excel.txt', 'r') as f:
    f_contents = f.readline() #this read the line in the contents() 
    print(f_contents) 
    f_contents = f.readline()
    print(f_contents) #this read the line (1st and 2nd line) in the contents() 


```

```python
with open('30days_Excel.txt', 'r') as f:
    f_contents = f.readline() #this read the line in the contents() 
    print(f_contents, end='') 
    f_contents = f.readline()
    f_contents = f.readline() #this read the line in the contents() 
    print(f_contents, end='') #this to cprevent extra spaces in between lines
    f_contents = f.readline()
```

```python
with open('30days_Excel.txt', 'r') as f:
    for line in f: #to get the lines without having to repeat the code
        print(line, end='')
```

```python
with open('30days_Excel.txt', 'r') as f:
    f_contents = f.readline(50) # this read the first 50 characters in the line in the contents() 
    print(f_contents, end='') 
```

### Write to files

```python
with open('30days_Excel2.txt', 'w') as f:
    pass #only creates a file without writing anything
```

```python
with open('30days_Excel2.txt', 'w') as f:
    f.write('Test') #writes the word 'Test' inside the file
    f.write('Test') # create mulitple without space
    
```

```python
with open('30days_Excel2.txt', 'w') as f:
    f.write('Test') #writes the word 'Test' inside the file
    f.seek(0) #diff from it usecase in reading, it returns the last command without adding the previous input
    f.write('Test') # create mulitple without space
    
```
